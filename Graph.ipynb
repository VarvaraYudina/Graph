# Построение графика

Импортируем библиотеки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Вводим переменные f - величина отклонения и n - количество точек на графике

```python
f = 5
n = 100
```

Вводим функции

```python
rlist1 = []
rlist2 = []
for i in np.linspace(0,3,n):
    rlist1.append({'par1':i,
                  'par2':np.sqrt(i),
                  'par3': np.sqrt(i)-(np.random.random() - 0.5)/f,
                  'par4':2*np.power(np.sin(i/2),2),
                  'par5':2*np.power(np.sin(i/2),2)-(np.random.random() - 0.5)/f})
    rlist2.append({'par1':i,
                  'par2':np.power(i,2)/8,
                  'par3': (np.power(i,2)/8-(np.random.random() - 0.5)/f),
                  'par4':(np.exp(i)-1)/10,
                  'par5':((np.exp(i)-1)/10-(np.random.random() - 0.5)/f)})
```

Создаем графики

```python
data1 = pd.DataFrame(rlist1)
data2 = pd.DataFrame(rlist2)

#Добавляем название, цвет, количество графиков и названия осей
chart_color = '#f5e4ce'
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.set_title("Зависимость скорости от диаметра панциря черепах")
ax2.set_xlabel("Диаметр")
ax1.set_ylabel("Скорость")
ax2.set_ylabel("Скорость")

#Убираем тики
ax1.tick_params(axis='x',which='both',bottom=False)
ax1.tick_params(axis='y',which='both',left=False)
ax2.tick_params(axis='x',which='both',bottom=False)
ax2.tick_params(axis='y',which='both',left=False)

#Добавляем цвет графика 
ax1.set_facecolor(chart_color)
ax2.set_facecolor(chart_color)

#Создаем легенду
ax1.plot(data1['par1'], data1['par2'], 'r-', label='Морские')
ax1.plot(data1['par1'], data1['par3'], 'r--')
ax1.plot(data1['par1'], data1['par4'], 'b-', label='Летающие')
ax1.plot(data1['par1'], data1['par5'], 'b--')

ax2.plot(data2['par1'], data2['par2'], 'r-', label='Морские')
ax2.plot(data2['par1'], data2['par3'], 'r--')
ax2.plot(data2['par1'], data2['par4'], 'b-', label='Летающие')
ax2.plot(data2['par1'], data2['par5'], 'b--')

legend=ax1.legend(loc='lower right')
frame = legend.get_frame()
frame.set_facecolor(chart_color)
frame.set_edgecolor(chart_color)

legend=ax2.legend(loc='lower right')
frame = legend.get_frame()
frame.set_facecolor(chart_color)
frame.set_edgecolor(chart_color)

ax1.text(0.2, 1.7,"В сферическом вакууме", style='italic')
ax2.text(0.2, 1.7,"В жидком азоте", style='italic')

plt.show()
```
